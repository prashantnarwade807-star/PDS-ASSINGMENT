Questions and Answers:

1. What is the difference between Git and GitHub?Git is a distributed version control system used to track changes in files and manage project versions locally on a computer. It allows developers to work on code, maintain history, and revert changes when needed. On the other hand, GitHub is a cloud-basedplatform that hosts Git repositories online and provides additional features like collaboration, sharing, issue tracking, and project management. In simple terms, Git is a tool, while GitHub is a service that uses Git.

2. Explain the difference between git add and git commit

The git add command is used to move changes from the working directory to the staging area, meaning it prepares the files to be committed. It does not save the changes permanently. The git commit command, however, records the staged changes into the repository along with a message describing the update. In short, git add selects the changes, while git commit saves them permanently.

3. What is the staging area in Git?

The staging area in Git is an intermediate space where changes are stored before they are committed to the repository. It allows users to review and organize changes by selecting which files or parts of files should be included in the next commit. This helps in maintaining clean and meaningful commit history by avoiding unnecessary or incomplete changes.

4. What happens if you modify a file after committing it?
If a file is modified after it has been committed, Git recognizes it as a new change and marks it as modified in the working directory. The previous committed version remains unchanged in the repository. To save the new changes, the user must again add the file to the staging area using git add and then commit it using git commit.

5. Difference between git push and git pull

The git push command is used to upload local repository changes to a remote repository such as GitHub, making them available to others. In contrast, the git pull command is used to fetch and download changes from the remote repository and merge them into the local repository. Essentially, git push sends data from local to remote, while git pull brings updates from remote to local.
