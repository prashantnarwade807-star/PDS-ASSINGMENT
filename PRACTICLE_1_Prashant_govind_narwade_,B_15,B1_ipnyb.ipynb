 1. What are sequence data types in Python?
   Name them and explain how indexing and slicing work across different sequence types?


ANS : **Sequence data types** in Python are ordered collections of items. The main ones are **strings**, **lists**, and **tuples**.

*   **Indexing** allows you to access individual elements by their position (e.g., `my_list[0]` for the first element, `my_string[-1]` for the last).
*   **Slicing** extracts a portion of a sequence using `[start:end:step]` (e.g., `my_tuple[1:4]` gets elements from index 1 up to, but not including, index 4).

2. What is the difference between a list and a tuple in Python?
   Why would you prefer a tuple over a list in real-world applications?
ANS:
*   **Lists** are **mutable** (changeable), defined with square brackets `[]`. Use them when your collection of items needs to be modified (add, remove, change elements).
*   **Tuples** are **immutable** (unchangeable), defined with parentheses `()`. Prefer tuples for data that should remain constant (e.g., coordinates, RGB colors), as function return values, or as dictionary keys (because of their immutability).

3. How are Python arrays different from lists?
   When should you use the array module instead of a list?
   
ANS:
*   **Lists** are heterogeneous (can store different data types) and general-purpose.
*   **`array.array`** (from the `array` module) are homogeneous (store only one specific data type) and are more memory-efficient and potentially faster for large collections of primitive numerical data. Use `array.array` for large numerical datasets, when interfacing with C/C++ libraries, or for efficient binary file I/O.

4. Explain how dictionaries work internally in Python.
   Why are dictionary lookups faster compared to lists or tuples?

  ANSWER :

Python dictionaries are implemented as **hash tables**.

*   When an item is added, a **hash value** is computed for the key, which determines where it's stored.
*   **Lookups are fast (average O(1))** because the hash value allows Python to directly jump to the probable location of the key-value pair, minimizing the need for sequential searching, unlike lists or tuples which require iterating through elements.

5. What is the difference between sets and lists in Python?
   How do sets handle duplicate values, and where are sets useful in practical scenarios?

   ANSWER:
   

*   **Lists** are ordered and can contain duplicates.
*   **Sets** are unordered and do **not allow duplicate** values. If you try to add an existing element, it's ignored. Sets are useful for:
    *   Efficiently **removing duplicates** from a collection.
    *   Very fast **membership testing** (`x in my_set`).
    *   Performing mathematical **set operations** like union, intersection, and difference.

```python
import array

print("--- a) Strings ---")
my_string = "Hello Python World"
print(f"Original string: {my_string}")
print(f"First character: {my_string[0]}")
print(f"Last character: {my_string[-1]}")
print(f"Slice (6 to 12): {my_string[6:12]}")
print(f"Length: {len(my_string)}")
print(f"Check for 'Python' in string: {'Python' in my_string}")
print(f"Uppercase: {my_string.upper()}")

print("\n--- b) Lists ---")
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")
print(f"Second element: {my_list[1]}")
my_list.append(60)
print(f"After append(60): {my_list}")
my_list.insert(1, 15)
print(f"After insert(1, 15): {my_list}")
my_list.remove(40)
print(f"After remove(40): {my_list}")
my_list[0] = 5
print(f"After changing first element: {my_list}")
print(f"Length: {len(my_list)}")
print(f"Check for 30 in list: {30 in my_list}")

print("\n--- c) Arrays (using array module) ---")
my_array = array.array('i', [10, 20, 30, 40, 50]) # 'i' for signed integer
print(f"Original array: {my_array}")
print(f"Third element: {my_array[2]}")
my_array.append(60)
print(f"After append(60): {my_array}")
my_array[0] = 5
print(f"After changing first element: {my_array}")
print(f"Length: {len(my_array)}")

print("\n--- d) Tuples ---")
my_tuple = (1, "apple", 3.14, True, 1)
print(f"Original tuple: {my_tuple}")
print(f"First element: {my_tuple[0]}")
print(f"Slice (1 to 3): {my_tuple[1:4]}")
print(f"Length: {len(my_tuple)}")
print(f"Count of 1: {my_tuple.count(1)}")
print(f"Index of 'apple': {my_tuple.index('apple')}")

print("\n--- e) Dictionaries ---")
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
print(f"Original dictionary: {my_dict}")
print(f"Name: {my_dict['name']}")
my_dict['age'] = 31
print(f"After updating age: {my_dict}")
my_dict['occupation'] = 'Engineer'
print(f"After adding occupation: {my_dict}")
print(f"Keys: {my_dict.keys()}")
print(f"Values: {my_dict.values()}")
print(f"Items: {my_dict.items()}")
print(f"Check for 'city' in keys: {'city' in my_dict}")

print("\n--- f) Sets ---")
my_set = {1, 2, 3, 2, 4, 1}
print(f"Original set (duplicates removed): {my_set}")
my_set.add(5)
print(f"After adding 5: {my_set}")
my_set.remove(2)
print(f"After removing 2: {my_set}")

set1 = {1, 3, 5}
set2 = {3, 5, 7}
print(f"Union of set1 and set2: {set1.union(set2)}")
print(f"Intersection of set1 and set2: {set1.intersection(set2)}")
print(f"Difference (set1 - set2): {set1.difference(set2)}")
print(f"Check for 3 in set: {3 in my_set}")

print("\n--- g) Range ---")
my_range = range(5) # Generates numbers from 0 to 4
print(f"Range(5) converted to list: {list(my_range)}")
my_range_start_stop = range(2, 7) # Generates numbers from 2 to 6
print(f"Range(2, 7) converted to list: {list(my_range_start_stop)}")
my_range_step = range(0, 10, 2) # Generates numbers from 0 to 9 with step 2
print(f"Range(0, 10, 2) converted to list: {list(my_range_step)}")
print(f"Check if 4 is in range(5): {4 in my_range}")
```
