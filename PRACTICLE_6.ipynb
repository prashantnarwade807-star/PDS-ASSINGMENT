### 1. What is web scraping? Explain the role of the Requests library in web scraping.
*   **Web scraping** is the automated process of extracting data from websites. Instead of manually copying data, web scraping tools use bots or scripts to browse websites and collect specific information, which is then typically stored in a structured format (like a CSV file, database, or JSON). It's often used for data analysis, market research, price comparison, content aggregation, and more.
*   The **Requests library** in Python is a fundamental tool for web scraping because it allows you to send HTTP requests to web servers. In web scraping, its primary role is to **fetch the HTML content of a webpage**. Before you can parse or extract data from a website, you need to download its content, and `Requests` handles this by allowing you to make `GET` requests (to retrieve data), `POST` requests (to send data), and other HTTP methods. It simplifies interaction with web services, handling complexities like redirects, sessions, and authentication.

### 2. What is BeautifulSoup? How does it help in parsing HTML?
*   **BeautifulSoup** is a Python library designed for parsing HTML and XML documents. It creates a parse tree from the page source code, which can then be used to extract data from the HTML in a very easy and intuitive way.
*   **How it helps in parsing HTML:** After the `Requests` library fetches the raw HTML content of a webpage, BeautifulSoup takes this raw HTML and transforms it into a tree-like structure of Python objects. This structure makes it very easy to navigate, search, and modify the parse tree. You can find specific HTML tags by name, by their attributes (like `class` or `id`), or by their text content. It handles malformed HTML gracefully, making it robust for real-world web pages.

### 3. Differentiate between web scraping and web crawling.
*   **Web Scraping:** Focuses on **extracting specific data** from a web page or a set of web pages. The goal is to obtain the data itself (e.g., product prices, article text, contact information). It typically starts with a known URL and then extracts content from that page.
*   **Web Crawling:** Focuses on **discovering and indexing web pages** by following hyperlinks from one page to another. The goal is to build a comprehensive map of a website or the entire internet. Search engines use web crawling to build their indices. A web crawler moves from page to page, and while it might perform some scraping on each page, its primary purpose is navigation and discovery rather than detailed data extraction.
*   **In summary:** Crawling is about *finding* pages, while scraping is about *getting data from* pages. They often work hand-in-hand: a crawler finds pages, and a scraper extracts data from them.

### 4. What are the commonly used methods in BeautifulSoup?
Some commonly used methods in BeautifulSoup for navigating and searching the parse tree include:
*   `BeautifulSoup(html_doc, 'html.parser')`: Initializes a BeautifulSoup object.
*   `find(name, attrs, string)`: Finds the first tag matching the given criteria.
*   `find_all(name, attrs, string, limit)`: Finds all tags matching the given criteria, returning a list. (Also accessible as `soup(name, attrs, string)`).
*   `select(selector)`: Allows you to find elements using CSS selectors (e.g., `soup.select('div.product-name')`).
*   `.text` or `.get_text()`: Extracts the visible text from a tag.
*   `.<tag_name>`: Accesses the first child tag with that name (e.g., `soup.p`).
*   `[attribute_name]`: Accesses an attribute of a tag (e.g., `link['href']`).
*   `.contents`, `.children`, `.descendants`: For navigating through child and descendant elements.
*   `.parent`, `.parents`: For navigating to parent elements.
*   `.next_sibling`, `.previous_sibling`: For navigating to sibling elements.

### 5. What are the ethical and legal considerations in web scraping?
Web scraping, while powerful, comes with significant ethical and legal considerations:
*   **Ethical Considerations:**
    *   **Respect `robots.txt`:** This file on a website tells crawlers and scrapers which parts of the site they are allowed or forbidden to access. Ethically, you should always respect these directives.
    *   **Server Load:** Scraping too aggressively can overload a server, leading to denial-of-service for legitimate users. Be mindful of request frequency and introduce delays (`time.sleep()`).
    *   **Data Usage:** Consider how the scraped data will be used. Is it for personal analysis, commercial purposes, or redistribution? Using scraped data in a way that competes unfairly or misrepresents the source is unethical.
    *   **Privacy:** Be extremely cautious when scraping personal data. Violating privacy laws (like GDPR, CCPA) can lead to severe penalties.
*   **Legal Considerations:**
    *   **Terms of Service (ToS):** Many websites have ToS that explicitly prohibit scraping. Violating ToS can lead to legal action, even if the data is publicly available.
    *   **Copyright:** Scraped content (text, images, videos) is often copyrighted. Republishing or commercially exploiting copyrighted material without permission is illegal.
    *   **Trespass to Chattels / Computer Fraud and Abuse Act (CFAA) in the US:** In some jurisdictions, aggressive scraping that causes damage or significant burden to a server can be considered a form of digital trespass or unauthorized access.
    *   **Data Protection Laws (GDPR, CCPA, etc.):** If you scrape personal data (e.g., names, email addresses), you must comply with relevant data protection laws regarding its collection, storage, and processing. This often requires consent.
    *   **Contract Law:** If you've entered into an agreement (even by clicking 'I Agree' on a website), you might be bound by its terms against scraping.

**Best practices:** Always check a website's `robots.txt`, review its Terms of Service, limit your request rate, avoid scraping personal or sensitive information, and ensure you have a legitimate, legal, and ethical reason for scraping the data.

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Define the URL of the webpage to scrape
url = "https://example.com"

# Step 2: Define headers to simulate a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

try:
    # Step 3: Send HTTP request to the server
    response = requests.get(url, headers=headers, timeout=10)

    # Step 4: Check if request is successful
    if response.status_code == 200:
        print("Webpage fetched successfully!\n")
    else:
        print("Failed to fetch webpage. Status Code:", response.status_code)

    # Step 5: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Step 6: Extract and display the title of the webpage
    if soup.title:
        print("Page Title:")
        print(soup.title.string, "\n")
    else:
        print("No title found\n")

    # Step 7: Extract all links (<a> tags)
    print("List of Links:\n")
    links = soup.find_all("a")

    link_count = 0
    for link in links:
        href = link.get("href")
        text = link.get_text(strip=True)

        if href:
            link_count += 1
            print(f"{link_count}. Text: {text if text else 'No text'}")
            print(f"   URL: {href}\n")

    # Step 8: Extract headings (h1, h2, h3)
    print("Headings:\n")

    for tag in ["h1", "h2", "h3"]:
        headings = soup.find_all(tag)

        if headings:
            print(f"--- {tag.upper()} Tags ---")
            for heading in headings:
                print(heading.get_text(strip=True))
            print()

    # Step 9: Extract paragraph data
    print("Paragraphs:\n")
    paragraphs = soup.find_all("p")

    para_count = 0
    for para in paragraphs:
        text = para.get_text(strip=True)
        if text:
            para_count += 1
            print(f"{para_count}. {text}\n")

    # Step 10: Extract image sources
    print("Image Sources:\n")
    images = soup.find_all("img")

    img_count = 0
    for img in images:
        src = img.get("src")
        if src:
            img_count += 1
            print(f"{img_count}. {src}")

    # Step 11: Display summary of extracted data
    print("\nSummary:")
    print("Total Links:", link_count)
    print("Total Headings:",
          len(soup.find_all('h1')) +
          len(soup.find_all('h2')) +
          len(soup.find_all('h3')))
    print("Total Paragraphs:", para_count)
    print("Total Images:", img_count)

# Step 12: Handle errors properly
except requests.exceptions.RequestException as e:
    print("Error in HTTP request:", e)

except Exception as e:
    print("Unexpected error occurred:", e)
```
