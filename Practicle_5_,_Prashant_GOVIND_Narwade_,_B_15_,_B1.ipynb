1.  **Importance of data visualization in data analysis and EDA:**
    *   **Understanding patterns:** Visualization helps in quickly identifying trends, patterns, and outliers in data that might be hidden in raw numerical form.
    *   **Communication:** It's an effective way to communicate complex findings and insights to a non-technical audience.
    *   **Decision making:** Clear visualizations can support better and faster decision-making.
    *   **In EDA:** Visualization is crucial for:
        *   **Identifying anomalies:** Spotting unusual data points or errors.
        *   **Detecting relationships:** Understanding correlations between variables.
        *   **Exploring distributions:** Seeing how data is spread across different values.
        *   **Formulating hypotheses:** Generating ideas for further analysis.

2.  **Difference between Matplotlib and other visualization libraries:**
    *   **Matplotlib:** It's a foundational, low-level plotting library in Python. It provides extensive control over every element of a plot, making it highly customizable. However, this also means it can be more verbose for complex plots.
    *   **Other libraries (e.g., Seaborn, Plotly, Altair):** These libraries are often built on top of Matplotlib or offer a higher-level API. They typically provide:
        *   **Easier syntax:** For creating common statistical plots with fewer lines of code.
        *   **Better aesthetics by default:** Often producing more visually appealing plots with less manual tweaking.
        *   **Specialized functionalities:** Seaborn, for example, excels at statistical plots, while Plotly and Altair are good for interactive visualizations.
        *   While Matplotlib offers fine-grained control, other libraries aim to simplify common plotting tasks and enhance readability/interactivity.

3.  **Purpose of the `pyplot` module in Matplotlib:**
    *   The `matplotlib.pyplot` module is a collection of functions that make Matplotlib work like MATLAB. It provides a convenient, high-level interface for creating plots.
    *   It automatically handles the creation of figures and axes to simplify the plotting process.
    *   Common functions like `plt.plot()`, `plt.scatter()`, `plt.xlabel()`, `plt.title()`, `plt.show()` all come from `pyplot`.
    *   It's the most commonly used interface for interactive plotting in Matplotlib.

4.  **Process of creating a basic line plot:**
    *   **Import `matplotlib.pyplot`:** `import matplotlib.pyplot as plt`
    *   **Prepare data:** Have two lists or arrays for your x-axis and y-axis values (e.g., `x = [1, 2, 3, 4]`, `y = [2, 4, 6, 8]`).
    *   **Create the plot:** Call `plt.plot(x, y)`.
    *   **Add labels and title (optional but recommended):** `plt.xlabel('X-axis Label')`, `plt.ylabel('Y-axis Label')`, `plt.title('My Line Plot')`.
    *   **Display the plot:** `plt.show()`.

5.  **Why customization is important and available options in Matplotlib:**
    *   **Importance of customization:**
        *   **Clarity and readability:** Making plots easy to understand.
        *   **Emphasis:** Highlighting key data points or trends.
        *   **Branding and consistency:** Matching visual styles for reports or presentations.
        *   **Accessibility:** Ensuring plots are perceivable by a wider audience (e.g., colorblind-friendly palettes).
        *   **Aesthetics:** Making plots visually appealing.
    *   **Customization options in Matplotlib:** Matplotlib offers an extensive range of customization, including:
        *   **Colors:** Line colors, marker colors, fill colors (`color='red'`, `cmap='viridis'`)
        *   **Line styles:** Solid, dashed, dotted (`linestyle='--'`)
        *   **Markers:** Circles, squares, triangles for data points (`marker='o'`)
        *   **Labels and titles:** `xlabel()`, `ylabel()`, `title()`
        *   **Legends:** `plt.legend()` to identify different data series.
        *   **Axes limits and ticks:** `xlim()`, `ylim()`, `xticks()`, `yticks()`
        *   **Fonts:** Size, style, weight of text elements.
        *   **Grids:** `plt.grid(True)` to add gridlines.
        *   **Figure size:** `plt.figure(figsize=(width, height))`
        *   **Subplots:** `plt.subplot()` for arranging multiple plots in a single figure.
        *   **Annotations and text:** `plt.text()`, `plt.annotate()` for adding specific notes.
        *   **Saving plots:** `plt.savefig('my_plot.png')` for specifying format and resolution.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create a sample dataset
np.random.seed(42)
data = {
    'Feature_A': np.random.rand(100) * 100, # Numerical
    'Feature_B': np.random.randn(100) * 15 + 50, # Numerical (normally distributed)
    'Feature_C': np.random.randint(1, 5, 100), # Categorical (integer representation)
    'Feature_D': np.random.choice(['Group 1', 'Group 2', 'Group 3'], 100), # Categorical
    'Time_Series': pd.date_range(start='2023-01-01', periods=100, freq='D'), # Date/Time
    'Value_TS': np.cumsum(np.random.randn(100) * 5) + 100 # Time-series value
}
df = pd.DataFrame(data)

# Convert 'Feature_C' to categorical type
df['Feature_C'] = df['Feature_C'].astype('category')

print("Sample DataFrame head:")
display(df.head())

print("\n--- Data Visualizations ---")
```

```python
# Scatter Plot: Relationship between two numerical features
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Feature_A', y='Feature_B', hue='Feature_D', data=df)
plt.title('Scatter Plot of Feature A vs Feature B')
plt.xlabel('Feature A')
plt.ylabel('Feature B')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Line Plot: Trend of a value over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Time_Series', y='Value_TS', data=df)
plt.title('Line Plot of Value Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Bar Plot: Distribution of a categorical feature
plt.figure(figsize=(8, 6))
sns.countplot(x='Feature_D', data=df, palette='viridis', hue='Feature_D', legend=False)
plt.title('Bar Plot of Feature D Distribution')
plt.xlabel('Feature D Categories')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Bar Plot: Comparison of means of a numerical feature by category
plt.figure(figsize=(8, 6))
sns.barplot(x='Feature_D', y='Feature_A', data=df, estimator=np.mean, palette='muted', hue='Feature_D', legend=False)
plt.title('Mean of Feature A by Feature D')
plt.xlabel('Feature D Categories')
plt.ylabel('Mean Feature A')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Histogram: Distribution of a single numerical feature
plt.figure(figsize=(8, 6))
sns.histplot(df['Feature_B'], bins=15, kde=True, color='skyblue')
plt.title('Histogram of Feature B')
plt.xlabel('Feature B Values')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Box Plot: Distribution and outliers of numerical features by category
plt.figure(figsize=(8, 6))
sns.boxplot(x='Feature_D', y='Feature_A', data=df, palette='pastel', hue='Feature_D', legend=False)
plt.title('Box Plot of Feature A by Feature D')
plt.xlabel('Feature D Categories')
plt.ylabel('Feature A Values')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Pair Plot: Relationships between all numerical features in the dataset
print("\nPair Plot: Visualizing relationships between numerical features...")
sns.pairplot(df[['Feature_A', 'Feature_B', 'Value_TS', 'Feature_D']], hue='Feature_D', palette='tab10')
plt.suptitle('Pair Plot of Numerical Features', y=1.02) # Adjust suptitle position
plt.show()
```
