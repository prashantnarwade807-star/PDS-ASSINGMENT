1. What is correlation? How is it different from regression?

Correlation is a statistical method used to measure the strength and direction of the relationship between two variables.

Positive correlation: both variables increase together.
Negative correlation: one increases while the other decreases.
Zero correlation: no relationship exists.

Example:

Height and weight usually have positive correlation.

Difference between Correlation and Regression

| Correlation                             | Regression                              |
| --------------------------------------- | --------------------------------------- |
| Measures relationship between variables | Predicts one variable from another      |
| Shows strength and direction            | Shows cause-and-effect type prediction  |
| No dependent or independent variable    | Has dependent and independent variables |
| Value lies between -1 and +1            | Produces an equation                    |
| Symmetric in nature                     | Not symmetric                           |


2. What is simple linear regression? Explain with suitable example.

Simple Linear Regression is a statistical technique used to model the relationship between one independent variable and one dependent variable using a straight line.

The regression equation is:

y=a+bx

Where:

y = dependent variable
x = independent variable
a = intercept
b = slope
Example

Suppose a teacher wants to predict student marks based on study hours.

Study Hours (x)	Marks (y)
1	30
2	40
3	50
4	60

The regression line may become:

y=20+10x

If a student studies 5 hours:

y=20+10(5)=70

So predicted marks are 70.

4. What is multicollinearity? Does it affect simple regression?

Multicollinearity occurs when independent variables in a regression model are highly correlated with each other.

Example:

Salary and years of experience may be highly related.
Effects of Multicollinearity
Regression coefficients become unstable.
Difficult to identify important variables.
Prediction accuracy may reduce.
Does it affect simple regression?

No.
Simple linear regression has only one independent variable, so multicollinearity does not occur.

It mainly affects multiple regression models.

5. A dataset shows strong correlation but poor prediction accuracy. Why?

A strong correlation does not always guarantee accurate predictions because:

Correlation only measures relationship
It does not ensure a perfect predictive model.
Non-linear relationship
Variables may be related non-linearly while regression assumes linearity.
Outliers
Extreme values can affect prediction accuracy.
Overfitting or underfitting
Model may not generalize well.
Missing important variables
Other factors affecting prediction may not be included.
Correlation is not causation
Two variables may move together without directly influencing each other.
Example

Ice cream sales and drowning incidents may show strong correlation because both increase during summer, but ice cream sales do not cause drowning.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import pearsonr

X = np.array([10, 20, 30, 40, 50])
Y = np.array([15, 25, 35, 45, 55])

def pearson_coefficient(x, y):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sqrt(np.sum((x - mean_x)*2) * np.sum((y - mean_y)*2))
    return numerator / denominator

r_manual = pearson_coefficient(X, Y)
r_scipy, _ = pearsonr(X, Y)

print("Karl Pearson Correlation")
print(f"Manual Result: {r_manual:.4f}")
print(f"SciPy Result: {r_scipy:.4f}")

if r_manual == 1:
    print("Perfect Positive Correlation")
elif r_manual > 0:
    print("Positive Correlation")
elif r_manual == 0:
    print("No Correlation")
else:
    print("Negative Correlation")

X_reshaped = X.reshape(-1, 1)

model = LinearRegression()
model.fit(X_reshaped, Y)

Y_pred = model.predict(X_reshaped)

m = model.coef_[0]
c = model.intercept_

print("\nLinear Regression")
print(f"Equation: Y = {m:.2f}X + {c:.2f}")

mse = mean_squared_error(Y, Y_pred)
r2 = r2_score(Y, Y_pred)

print(f"MSE: {mse:.4f}")
print(f"R2 Score: {r2:.4f}")

plt.scatter(X, Y, label='Actual Data')
plt.plot(X, Y_pred, label='Regression Line')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Correlation and Linear Regression")
plt.legend()
plt.grid(True)
plt.show()

print("\nConclusion:")
print("Strong relationship between X and Y")
print("Perfect positive correlation (r = 1)")
print("Model fits data perfectly")
```
